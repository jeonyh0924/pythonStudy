# 양방향 연결 리스트 

```python
class Node:
    def __init__(self, value, n= None, prev= None):
        self.value = value
        self.n = n
        self.prev = prev
        
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
    
    def is_empty(self):
        # 리스트가 비어있는지 확인하는 연산자
        node_ins_exists = self.head or self.tail
        if node_ins_exists:
            return False
        else:
            return True

    def prepend(self, value):
        # 리스트의 앞에 개체를 삽입(prepending)하는 연산자
        past_head = self.head
        
        
        self.head = Node(value, n=self.head)
        if past_head:
            past_head.prev = self.head
            return f'prepend: {self.head.value} next is {past_head.value}'
        else:
            self.tail = self.head
            return f'prepend: prev, tail is {self.head.value}!'

    def append(self, value):
        # 리스트의 뒤에 개체를 삽입(appending)하는 연산자
        past_tail = self.tail
        
        self.tail = Node(value, prev=self.tail)
        
        if past_tail:
            past_tail.n = self.tail
            return f'append: {self.tail.value} prev is {past_tail.value}'
        else:
            self.head = self.tail
            return f'append: prev, tail is {self.tail.value}'

    def set_head(self, index):
        # 리스트의 첫 머리(head)를 결정하는 연산자
        obj = self.head
        
        for i in range(index):
            obj = obj.n
        obj.prev = None
        self.head = obj
        return f'set_head: {self.head.value} is head'
        

    def access(self, index):
        # 주어진 인덱스에 해당하는 요소에 접근하는 연산자
        obj = self.head
        
        for i in range(index):
            obj = obj.n
        return f'{obj.value}'

    def insert(self, index, value):
        # 주어진 인덱스에 새로운 요소를 삽입하는 연산자
        obj = self.head
        
        for i in range(index):
            obj = obj.n
        node_obj = Node(value, n=obj, prev=obj.prev)
        
        obj.prev.n = node_obj
        
        return f'insert: {obj.value} prev is {node_obj.value} '

    def remove(self, index):
        # 주어진 인덱스에 해당하는 요소를 제거하는 연산자
        obj = self.head
        
        for i in range(index):
            obj = obj.n
        obj.prev.n = obj.n
        obj.n.prev = obj.prev
        return f'remove: {obj.value} is delete {obj.prev.value} next is {obj.prev.n.value}, {obj.n.value} prev is {obj.prev.value}'

    def print(self):
        head = self.head
        while True:
            if head:
                print(f'{head.value}', end=' ')
            else:
                print(f'')
                break
            head = head.n


d1 = DoublyLinkedList()
print(f'새로운 양방향 연결 리스트를 생성할 때 ins.is_empty()는', d1.is_empty())

d1.prepend(10)
d1.prepend(5)
d1.append(15)
d1.append(20)

d1.prepend(0)

print(d1.set_head(1))
d1.print()
print(d1.insert(2, 13))

d1.print()

print(d1.remove(2))
d1.print()
```
